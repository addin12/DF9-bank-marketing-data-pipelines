```python
# calculate the number of values to mask
column_length = len(df[['job','marital','education','default','housing',
                        'loan','contact','month','day_of_week','poutcome','y']])

num_values_to_mask = int(0.001 * column_length) 
```

```python
# generate a random boolean mask with the same shape as the 'age' column
mask = np.random.rand(column_length, 11) < num_values_to_mask / column_length

# mask the selected values with null values
df.loc[:, ['job','marital','education','default','housing',
            'loan','contact','month','day_of_week','poutcome','y']]= df.loc[:, ['job','marital','education','default','housing',
            'loan','contact','month','day_of_week','poutcome','y']].mask(mask)
```
